```python
import pandas as pd
import scipy.stats
import math
```

```python
# Load and preprocess data
df = pd.read_csv('Kohli.csv')
df = df[df['Dismissal'] != 'not out']
df = df[df['Dismissal'] != '-']

#change the data type of runs to int
df['Runs'] = df['Runs'].astype(int) 
df
```

```python
n = 30
alpha = 0.05
h0 = df['Runs'].mean() # Null Hypothesis --> Average is equal to 43.23
samp_data = df.sample(n)
```

```python
h1 = samp_data['Runs'].mean() # Alternate Hypothesis --> Avg is different
samp_SD = samp_data['Runs'].std()
SE = samp_SD/math.sqrt(n)
t = (H1-H0)/SE
p = scipy.stats.t.sf(abs(t), df=n-1)*2
```

```python
print('p-value = ' + str(round(p, 2)))
print('p-value = ' + str(round(p*100, 2)) + '%')

if p < alpha:
    print('Reject the Null Hypothesis as the p-value of sample data is less than the 5% (0.05) level of significance')
else:
    print('Do not reject the Null Hypothesis. P-value is greater than or equal to alpha (5% or 0.05)')
```
